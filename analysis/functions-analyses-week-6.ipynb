```python
def numpydata_from_csv(csv_file):
    tmp_data = pandas.read_csv(csv_file, header =None) 
    data = tmp_data.to_numpy()
    return data
```

```python
import numpy as np
def CO2_Emissions():
    sum_data = "EEB C177 Revised Data Set .csv"
    np.array(sum_data).astype(np.str)
    next(sum_data)
    for column in sum_data:  
        CO2_Emissions_total = numpy.sum(column[2])
        return CO2_Emissions_total
```

```python
CO2_Emissions()
```

```python
def plot_year_vs_total_CO2_emissions(name, CO2_Emissions):
    plt.plot([0.5, 1, 10],[CO2_Emissions])
    plt.axis([0, 12, 0, 100])
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions')
    plt.title('CO2 Emissions By Year' % name)
    plt.show()
    return
               
#Tried to figure our what the 'EOL while scanning string literal' error was but wasn't sure how to fix it. W
```
